# 1. Odtworzenie rysunku z page_3_43.pdf

```python
import numpy as np
import matplotlib.pyplot as plt

# Definiowanie funkcji f_3(t) jako odwrócony ząb piły
def f_3(t):
    t = t % (2 * np.pi)  # Okres 2pi
    if t < np.pi:
        return (np.pi / 2) - t
    else:
        return t - (3 * np.pi / 2)

# Generowanie wartości t i f_3(t)
t_values = np.linspace(-4 * np.pi, 4 * np.pi, 1000)
f_values = np.vectorize(f_3)(t_values)

# Rysowanie wykresu
plt.figure(figsize=(12, 6))
plt.plot(t_values, f_values, label='$f_3(t)$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.xlabel('$t$')
plt.ylabel('$f_3(t)$')
plt.ylim(-np.pi, np.pi)  # Skala osi Y
plt.xlim(-4 * np.pi, 4 * np.pi)  # Skala osi X
plt.gca().set_aspect('equal', adjustable='box')  # Ustawienie równych proporcji osi
plt.title('Wykres funkcji $f_3(t)$')
plt.grid(True)
plt.legend()
plt.show()
```

# 2. Wartości obliczone analitycznie i porównanie wyników z kodem źródłowym w programie Octave

```python
import numpy as np
import matplotlib.pyplot as plt

# Funkcja obliczająca współczynniki rozkładu
def pila3_17():
    cc = np.zeros(9)
    for i in range(9):
        cc[i] = (4 / np.pi) * (1 / (2*i + 1)**2)

    coeff = np.zeros(11)
    coeff[2] = cc[0]
    coeff[4] = cc[1]
    coeff[6] = cc[2]
    coeff[8] = cc[3]
    coeff[10] = cc[4]
    return coeff

# Wywołanie funkcji i zapisanie wartości współczynników
x = pila3_17()

# Wyświetlenie analitycznie obliczonych wartości współczynników
print('Analitycznie obliczone wartości współczynników rozkładu:')
for i in range(1, 11):
    print(x[i])

# Wykres analitycznie obliczonych wartości współczynników
plt.figure()
plt.plot(range(1, 11), x[1:], '*')  # Poprawione indeksowanie
plt.ylim(0, 1.5)  # Zwiększenie zakresu osi Y
plt.grid(True)
plt.title('Analitycznie obliczone wartości współczynników rozkładu')
plt.xlabel('Numer współczynnika')
plt.ylabel('Wartość współczynnika')
plt.show()
```

# 3.  Wartości obliczone numerycznie

```python
import numpy as np
import matplotlib.pyplot as plt

# Funkcja numerycznego obliczania współczynników rozkładu
def calculate_numerical_coefficients(N):
    coefficients = np.zeros(11)
    for n in range(11):
        b = 0
        for m in range(1, N + 1):
            t = ((m - 1) * np.pi) / (N / 2)
            if t <= np.pi:
                f = (np.pi / 2) - t
            if t >= np.pi:
                f = t - (3 * np.pi / 2)
            b += f * np.cos(n * t)
        b /= (N / 2)
        coefficients[n] = b
    return coefficients

# Wartości częstotliwości próbkowania
sampling_frequencies = [1, 2, 3, 5, 10, 12, 15, 20, 50, 100]

# Analitycznie obliczone wartości współczynników
analytical_coefficients = np.array([0.0, 1.2732395447351628, 0.0, 0.14147106052612918,
                                    0.0, 0.050929581789406514, 0.0, 0.025984480504799238,
                                    0.0, 0.015719006725125464, 0.0])

# Wygenerowanie osobnego wykresu dla każdej wartości częstotliwości próbkowania
for frequency in sampling_frequencies:
    numerical_coefficients = calculate_numerical_coefficients(frequency)
    print(f'Wartości numerycznych współczynników dla częstotliwości próbkowania {frequency} Hz:')
    print(numerical_coefficients)
    plt.figure(figsize=(8, 5))
    plt.plot(range(11), numerical_coefficients, label=f'Numeryczne, {frequency} Hz')
    plt.plot(range(11), analytical_coefficients, 'k--', label='Analityczne')
    plt.title(f'Wartości współczynników dla częstotliwości próbkowania {frequency} Hz')
    plt.xlabel('Numer współczynnika')
    plt.ylabel('Wartość współczynnika')
    plt.legend()
    plt.grid(True)
    plt.show()

# Wygenerowanie wspólnego wykresu dla każdej wartości częstotliwości próbkowania
plt.figure(figsize=(10, 8))
for frequency in sampling_frequencies:
    numerical_coefficients = calculate_numerical_coefficients(frequency)
    print(f'Wartości numerycznych współczynników dla częstotliwości próbkowania {frequency} Hz:')
    print(numerical_coefficients)
    plt.plot(range(11), numerical_coefficients, label=f'Numeryczne, {frequency} Hz')
plt.plot(range(11), analytical_coefficients, 'k--', label='Analityczne')
plt.title('Porównanie wartości współczynników dla różnych częstotliwości próbkowania')
plt.xlabel('Numer współczynnika')
plt.ylabel('Wartość współczynnika')
plt.legend()
plt.grid(True)
plt.show()
```

# 4. Opis i interpretacja.
W niniejszym raporcie analizuję porównanie wyników analitycznych i numerycznych dla zestawu współczynników rozkładu. Wartości analityczne zostały obliczone na podstawie wzoru analitycznego, natomiast wartości numeryczne zostały wygenerowane za pomocą algorytmu numerycznego dla różnych częstotliwości próbkowania.

    Wartości analityczne współczynników rozkładu są następujące:
      0.0, 1.2732395447351628, 0.0, 0.14147106052612918, 0.0, 0.050929581789406514, 0.0, 0.025984480504799238, 0.0, 0.015719006725125464.

Przedstawione wartości numeryczne zostały obliczone dla kolejnych częstotliwości próbkowania.     

* Częstotliwość próbkowania 1 Hz:        Wartości numerycznych współczynników są znacznie wyższe niż wartości analityczne dla wszystkich współczynników. Wynika to prawdopodobnie z niskiej częstotliwości próbkowania, co sprawia, że algorytm numeryczny nie jest w stanie dokładnie przybliżyć funkcji pierwotnej.

* Częstotliwość próbkowania 2 Hz: Numeryczne wartości współczynników są również wyższe niż wartości analityczne, ale już niższe niż dla 1 Hz. Nadal jednak występują znaczne różnice między wartościami analitycznymi a numerycznymi.

* Częstotliwość próbkowania 3 Hz:        Dla 3 Hz, numeryczne wartości współczynników zaczynają być bardziej zbliżone do wartości analitycznych, chociaż wciąż występują pewne różnice.

* Częstotliwość próbkowania 5 Hz:        Wartości numerycznych współczynników nadal różnią się od wartości analitycznych, ale różnice te są już mniejsze niż dla niższych częstotliwości próbkowania.

* Częstotliwość próbkowania 10 Hz:        Dla 10 Hz, niektóre numeryczne wartości współczynników zaczynają być zbliżone do wartości analitycznych, ale nadal występują różnice.

* Częstotliwość próbkowania 12 Hz:        Wyniki są zbliżone do tych dla 10 Hz, gdzie niektóre numeryczne wartości współczynników są zbliżone do wartości analitycznych.

*  Częstotliwość próbkowania 15 Hz:        Numeryczne wartości współczynników są bardziej zbliżone do wartości analitycznych, ale nadal występują pewne różnice, zwłaszcza dla nieparzystych współczynników.

* Częstotliwość próbkowania 20 Hz:        Wartości numerycznych współczynników są coraz bardziej zbliżone do wartości analitycznych, zwłaszcza dla parzystych współczynników.

* Częstotliwość próbkowania 50 Hz:        Dla 50 Hz, większość numerycznych wartości współczynników jest już zbliżona do wartości analitycznych, co sugeruje poprawność działania algorytmu numerycznego.

* Częstotliwość próbkowania 100 Hz:        Wartości numerycznych współczynników są bardzo zbliżone do wartości analitycznych, co sugeruje, że algorytm numeryczny działa prawidłowo.

Warto zauważyć, że wraz ze wzrostem częstotliwości próbkowania numeryczne wartości współczynników coraz bardziej przypominają wartości analityczne, co sugeruje, że wyższa częstotliwość próbkowania umożliwia lepszą aproksymację funkcji pierwotnej i dokładniejsze obliczenia współczynników.


Wyniki numeryczne, uzyskane przy różnych częstotliwościach próbkowania, są stopniowo zbliżane do wartości analitycznych w miarę zwiększania częstotliwości próbkowania. Możemy zauważyć, że dla bardzo niskich częstotliwości próbkowania (1 Hz, 2 Hz) numeryczne wartości współczynników znacznie odbiegają od wartości analitycznych. Wynika to prawdopodobnie z ograniczeń algorytmu numerycznego, który przy niskiej częstotliwości próbkowania nie jest w stanie dokładnie przybliżyć funkcji pierwotnej.

Jednakże, wraz ze wzrostem częstotliwości próbkowania, numeryczne wartości współczynników coraz bardziej zbliżają się do wartości analitycznych. Dla częstotliwości próbkowania powyżej 50 Hz, różnice między numerycznymi a analitycznymi wartościami są już minimalne. Wynika z tego, że wyższa częstotliwość próbkowania umożliwia lepszą aproksymację funkcji pierwotnej i dokładniejsze obliczenia współczynników.
